```python
import torch.utils.data.dataset as Dataset

class CSVDataset_DigestPath(Dataset):    
    def __init__(self, dataroot, csv_path, transforms_eval):
        self.dataroot = dataroot
        self.csv = pd.read_csv(csv_path)
        self.transforms = transforms_eval
        
    def __getitem__(self, index):
        img_fname = self.csv['img_fname'][index][:-4]
        label = self.csv['label'][index]
        if label == 1:
            img = Image.open(os.path.join(self.dataroot, 'Positive', self.csv['img_fname'][index]))
            mask = Image.open(os.path.join(self.dataroot, 'Positive_Mask', self.csv['mask_fname'][index]))
            mask = transforms.ToTensor()(mask)
        else:
            img = Image.open(os.path.join(self.dataroot, 'Negative', self.csv['img_fname'][index]))
            mask = torch.Tensor([torch.nan]).unsqueeze(dim=0)
        return self.transforms(img), mask, label, img_fname
    
    def __len__(self):
        return self.csv.shape[0]
```

```python
def save_hdf5(output_path, asset_dict, attr_dict= None, mode='w'):
    file = h5py.File(output_path, mode)
    for key, val in asset_dict.items():
        data_shape = val.shape
        if key not in file:
            data_type = val.dtype
            chunk_shape = (1, ) + data_shape[1:]
            maxshape = (None, ) + data_shape[1:]
            dset = file.create_dataset(key, shape=data_shape, maxshape=maxshape, chunks=chunk_shape, dtype=data_type)
            dset[:] = val
            if attr_dict is not None:
                if key in attr_dict.keys():
                    for attr_key, attr_val in attr_dict[key].items():
                        dset.attrs[attr_key] = attr_val
        else:
            dset = file[key]
            dset.resize(len(dset) + data_shape[0], axis=0)
            dset[-data_shape[0]:] = val
    file.close()
    return output_path

def id_collate(batch):
    new_batch = []
    ids = []
    for _batch in batch:
        new_batch.append(_batch[:-1])
        ids.append(_batch[-1])
    return default_collate(new_batch), ids

def prepare_img_tensor(img: torch.Tensor, patch_size=256):
    make_divisble = lambda l, patch_size: (l - (l % patch_size))
    b, c, w, h = img.shape
    load_size = make_divisble(w, patch_size), make_divisble(h, patch_size)
    w_4k, h_4k = w // patch_size, h // patch_size
    return transforms.CenterCrop(load_size)(img), w_4k, h_4k


def roll_batch2img(batch: torch.Tensor, w: int, h: int, patch_size=256):
    """
    Args:
        batch: Length x Channel x W x H
        
    Return:
        Image.PIL: (W' x H' X 3)
    """
    batch = batch.reshape(w, h, 3, patch_size, patch_size)
    img = rearrange(batch, 'p1 p2 c w h-> c (p1 w) (p2 h)').unsqueeze(dim=0)
    return Image.fromarray(tensorbatch2im(img)[0])

def eval_transforms(pretrained=False, patch_size=256):
    if pretrained:
        mean, std = (0.485, 0.456, 0.406), (0.229, 0.224, 0.225)
    else:
        mean, std = (0.5,0.5,0.5), (0.5,0.5,0.5)
        
    
    trnsfrms_val = transforms.Compose([transforms.ToTensor(), 
                                       transforms.Normalize(mean = mean, std = std)])
    return trnsfrms_val

def create_UMAP(library_path, save_path, dataset, enc_name, n=15, d=0.1):
    path = os.path.join(library_path, '%s_%s.pkl' % (dataset, enc_name))
    with open(path, 'rb') as handle:
        asset_dict = pickle.load(handle)
        embeddings, labels = asset_dict['embeddings'], asset_dict['labels']
        print(enc_name, embeddings.shape)
        if 'crc100k' in dataset:
            labels[labels=='MUS'] = 'STR'

    mapper = umap.UMAP(n_neighbors=n, min_dist=d).fit(embeddings)
    fig = plt.figure(figsize=(10, 10), dpi=100)
    umap.plot.points(mapper, labels=labels, width=600, height=600)
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, '%s_%s_umap_n%d_d%0.2f.jpg' % (dataset, enc_name, n, d)))


def create_region_embeddings(embeddings_dir, enc_name, dataset, save_patches=False, sprite_dim=128, 
                             patch_datasets='path/to/patch/datasets', assets_dir ='./ckpts/',
                             disentangle=-1, stage=-1):
    
    print("Extracting Features for '%s' via '%s'" % (dataset, enc_name))
    device256, device4k = torch.device('cuda:0'), torch.device('cuda:1')

    if enc_name == 'resnet50trunc_mean':
        model256 = resnet50_trunc_baseline(pretrained=True)
        eval_t = eval_transforms(pretrained=True)
        model256 = model256.to(device256)
        model4k = None
    elif enc_name == 'vit256mean':
        pretrained_weights256 = '../Checkpoints/vit256_small_dino.pth'
        model256 = get_vit256(pretrained_weights=pretrained_weights256)
        eval_t = eval_transforms(pretrained=False)
        model4k = None
    elif enc_name == 'vit4k' or enc_name == 'vit256mean_4k':
        pretrained_weights256 = '../Checkpoints/vit256_small_dino.pth'
        pretrained_weights4k = '../Checkpoints/vit4k_xs_dino.pth'
        model256 = get_vit256(pretrained_weights=pretrained_weights256)
        model4k = get_vit4k(pretrained_weights=pretrained_weights4k)
        eval_t = eval_transforms(pretrained=False)
    else:
        pass

    model256.eval()
    if model4k != None:
        model4k.eval()

    if dataset == 'iciarbach_raw':
        ### Train
        dataroot = os.path.join(patch_datasets, 'ICIAR2018/Train_Regions_Raw')
        dataset = torchvision.datasets.ImageFolder(dataroot, transform=eval_t)
        dataloader = torch.utils.data.DataLoader(dataset, batch_size=1, shuffle=False, num_workers=4)
        fname = os.path.join(embeddings_dir, 'iciarbach_raw_train_%s%s' % (enc_name, _stage))
        save_region_embeddings(model256=model256, model4k=model4k, fname=fname, dataloader=dataloader, 
                               dataset=dataset, save_patches=save_patches, 
                               sprite_dim=sprite_dim, is_imagefolder=True, concat_256_4k=concat_256_4k)
        
    elif dataset == 'iciarbach_sn':
        ### Train
        dataroot = os.path.join(patch_datasets, 'ICIAR2018/Train_Regions_SN')
        dataset = torchvision.datasets.ImageFolder(dataroot, transform=eval_t)
        dataloader = torch.utils.data.DataLoader(dataset, batch_size=1, shuffle=False, num_workers=4)
        fname = os.path.join(embeddings_dir, 'iciarbach_sn_train_%s%s' % (enc_name, _stage))
        save_region_embeddings(model256=model256, model4k=model4k, fname=fname, dataloader=dataloader, 
                               dataset=dataset, save_patches=save_patches, 
                               sprite_dim=sprite_dim, is_imagefolder=True, concat_256_4k=concat_256_4k)
    
    elif dataset == 'digestpath_lesionseg':
        ### Train
        dataroot = os.path.join(patch_datasets, 'MICCAI2019_DigestPath/Colonoscopy_tissue_segment_dataset/Train/')
        dataset = CSVDataset_DigestPath(
            dataroot=dataroot,
            csv_path=os.path.join(dataroot, 'train.csv'),
            transforms_eval=eval_t
        )
        dataloader = torch.utils.data.DataLoader(dataset, batch_size=1, shuffle=False, num_workers=4, collate_fn=id_collate)
        dataset_pklpath = os.path.join(embeddings_dir, f'digestpath_lesionseg_train_{enc_name}.pkl')
        if 'vit' in enc_name:
            cached_embeddings_h5path = os.path.join(embeddings_dir, 'digestpath_lesionseg_train_vit.pkl')
        else:
            cached_embeddings_h5path = os.path.join(embeddings_dir, f'digestpath_lesionseg_train_{enc_name}.pkl')
        
        save_mask_region_embeddings(
            dataset_pklpath=dataset_pklpath, cached_embeddings_h5path=cached_embeddings_h5path,
            model256=model256, model4k=model4k, fname=fname, dataloader=dataloader, 
            save_patches=save_patches, sprite_dim=sprite_dim, is_imagefolder=False)


def save_region_embeddings(
    dataset_pklpath: str, cached_embeddings_h5path: str,
    model256: torch.nn.module, model4k: torch.nn.module, dataloader,
    save_patches=False, sprite_dim=128, overwrite=False, is_imagefolder=False):

    
    if os.path.isfile(dataset_pklpath) and (overwrite == False):
        return None
    
    embeddings, labels, patches = [], [], []

    for batch_256, target in tqdm(dataloader):
        ### Prepare batch_256 to be divisble by 256
        assert (len(batch_256.shape) == 4) and (batch_256.shape[0] == 1)
        batch_256, w_4k, h_4k = prepare_img_tensor(batch_256)
        
        if save_patches:
            for img in batch_256:
                patches.append(tensor2im(input_image=img).resize(sprite_dim))

        with torch.no_grad():
            batch_256 = batch_256.unfold(2, 256, 256).unfold(3, 256, 256)
            batch_256 = rearrange(batch_256, 'b c p1 p2 w h -> (b p1 p2) c w h')
            print(batch_256.shape)

            ### Filtering Patches
            patch_filter_params = patch_filter_params={'isWhitePatch': {'satThresh': 5}, 
                                           'isBlackPatch': {'rgbThresh': 40}}
            to_filter = np.array([filterPatch(img) for img in tensorbatch2im(batch_256)])
            contains_tissue = 1-to_filter
            
            features_256 = []
            for mini_bs in range(0, batch_256.shape[0], 256):
                minibatch_256 = batch_256[mini_bs:mini_bs+256].to(device256, non_blocking=True)
                features_256.append(model256(minibatch_256.to(device256)).detach().cpu())
                
            features_256 = torch.vstack(features_256)
            features_256_mean = features_256[np.where(contains_tissue)].mean(dim=0).unsqueeze(dim=0)

            if model4k != None:
                features_256 = features_256.reshape(w_4k, h_4k, 384).transpose(0,1).transpose(0,2).unsqueeze(dim=0)
                features_256 = features_256.to(device4k, non_blocking=True)
                features_4k = model4k.forward(features_256).detach().cpu()
                if concat_256_4k:
                    features = torch.cat([features_256_mean, features_4k], dim=1)
                else:
                    features = features_4k
            else:
                features = features_256_mean

            embeddings.append(features)
            labels.append(target)

    embeddings = np.vstack(embeddings)
    labels = np.vstack(labels).squeeze()
    
    if is_imagefolder:
        id2label = dict(map(reversed, dataset.class_to_idx.items()))
        labels = np.array(list(map(id2label.get, labels.ravel())))

    asset_dict = {'embeddings': embeddings, 'labels': labels}
    if save_patches:
        asset_dict.update({'patches': patches})
    with open('%s.pkl' % (fname), 'wb') as handle:
        pickle.dump(asset_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
        
        
def save_mask_region_embeddings(
    dataset_pklpath: str, cached_embeddings_h5path: str,
    model256, model4k, fname, dataloader, dataset=None, is_imagefolder=False, 
    save_patches=False, sprite_dim=128, overwrite=False, concat_256_4k=False):
    
    if os.path.isfile('%s.pkl' % fname) and (overwrite == False):
        return None

    os.makedirs(fname, exist_ok=True)
    embeddings, labels = [], []
    patches = []
    
    for _, img_fname in tqdm(dataloader):
        batch_256, mask_256, target = _
        img_fname = img_fname[0]
        
        #if os.path.isfile(os.path.join(fname, img_fname+'.h5')):
        #    h5 = h5py.File(os.path.join(fname, img_fname+'.h5'), 'rb')
        #    features = h5['feature_256']

        with torch.no_grad():
            ### Prepare batch_256 to be divisble by 256
            assert (len(batch_256.shape) == 4) and (batch_256.shape[0] == 1)
            batch_256, w_4k, h_4k = prepare_img_tensor(batch_256)
            if not mask_256.isnan().any().item():
                mask_256, w_4k, h_4k = prepare_img_tensor(mask_256)
            
            batch_256 = batch_256.unfold(2, 256, 256).unfold(3, 256, 256)
            batch_256 = rearrange(batch_256, 'b c p1 p2 w h -> (b p1 p2) c w h')
            
            if mask_256.isnan().any().item():
                mask_256 = np.zeros(batch_256.shape[0])
            else:
                mask_256 = mask_256.unfold(2, 256, 256).unfold(3, 256, 256)
                mask_256 = rearrange(mask_256, 'b c p1 p2 w h -> (b p1 p2) c w h')
                mask_256 = np.array(mask_256.sum(dim=2).sum(dim=2)[:,0] / (256**2))

            ### Filtering Patches
            patch_filter_params = patch_filter_params={'isWhitePatch': {'satThresh': 5}, 
                                           'isBlackPatch': {'rgbThresh': 40}}
            to_filter = np.array([filterPatch(img) for img in tensorbatch2im(batch_256)])
            contains_tissue = 1-to_filter
            
            features_256 = []
            for mini_bs in range(0, batch_256.shape[0], 256):
                minibatch_256 = batch_256[mini_bs:mini_bs+256].to(device256, non_blocking=True)
                features_256.append(model256(minibatch_256.to(device256)).detach().cpu())
                
            features_256 = torch.vstack(features_256)
            features_256_mean = features_256[np.where(contains_tissue)].mean(dim=0).unsqueeze(dim=0)

            if model4k != None:
                features_256 = features_256.reshape(w_4k, h_4k, 384).transpose(0,1).transpose(0,2).unsqueeze(dim=0)
                features_256 = features_256.to(device4k, non_blocking=True)
                features_4k = model4k.forward(features_256).detach().cpu()
                if concat_256_4k:
                    features = torch.cat([features_256_mean, features_4k], dim=1)
                else:
                    features = features_4k
                
                asset_dict = {
                    'features_256': features_256.detach().cpu().numpy(),
                    'features_4k': features_4k.numpy(),
                    'mask_256': mask_256,
                    'contains_tissue': contains_tissue,
                    'label': np.array(target)
                }
                save_hdf5(os.path.join(fname, img_fname+'.h5'), asset_dict, mode='w')
            else:
                features = features_256_mean
                asset_dict = {
                    'features_256': features_256.numpy(),
                    'mask_256': mask_256,
                    'contains_tissue': contains_tissue,
                    'label': np.array(target)
                }
                save_hdf5(os.path.join(fname, img_fname+'.h5'), asset_dict, mode='w')

        embeddings.append(features)
        labels.append(target)

    embeddings = np.vstack(embeddings)
    labels = np.vstack(labels).squeeze()
    
    if is_imagefolder:
        id2label = dict(map(reversed, dataset.class_to_idx.items()))
        labels = np.array(list(map(id2label.get, labels.ravel())))

    asset_dict = {'embeddings': embeddings, 'labels': labels}
    if save_patches:
        asset_dict.update({'patches': patches})
    with open('%s.pkl' % (fname), 'wb') as handle:
        pickle.dump(asset_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
class HIPT(torch.nn.Module):
    def __init__(self, model256_path, model4k_path, device256=torch.device('cuda:0'), device4k=torch.device('cuda:1'),
                 patch_filter_params={'isWhitePatch': {'satThresh': 5}, 
                                      'isBlackPatch': {'rgbThresh': 40}}):
        super().__init__()
        self.model256 = get_vit256(pretrained_weights=model256_path).to(device256)
        self.model4k = get_vit4k(pretrained_weights=model4k_path).to(device4k)
        self.device256 = device256
        self.device4k = device4k
        self.patch_filter_params = patch_filter_params
    
    def forward(self, x):
        batch_256, w_256, h_256 = self.prepare_img_tensor(x)
        batch_256 = batch_256.unfold(2, 256, 256).unfold(3, 256, 256)
        batch_256 = rearrange(batch_256, 'b c p1 p2 w h -> (b p1 p2) c w h')
        
        features_256 = []
        for mini_bs in range(0, batch_256.shape[0], 256):
            minibatch_256 = batch_256[mini_bs:mini_bs+256].to(self.device256, non_blocking=True)
            features_256.append(self.model256(minibatch_256).detach().cpu())

        features_256 = torch.vstack(features_256)
        features_256 = features_256.reshape(w_256, h_256, 384).transpose(0,1).transpose(0,2).unsqueeze(dim=0)
        features_256 = features_256.to(self.device4k, non_blocking=True)
        features_4k = self.model4k.forward(features_256)
        return features_4k
        
    def forward_asset_dict(self, x: torch.Tensor):
        batch_256, w_256, h_256 = self.prepare_img_tensor(x)
        batch_256 = batch_256.unfold(2, 256, 256).unfold(3, 256, 256)
        batch_256 = rearrange(batch_256, 'b c p1 p2 w h -> (b p1 p2) c w h')
        
        features_256 = []
        for mini_bs in range(0, batch_256.shape[0], 256):
            minibatch_256 = batch_256[mini_bs:mini_bs+256].to(self.device256, non_blocking=True)
            features_256.append(self.model256(minibatch_256).detach().cpu())

        features_256 = torch.vstack(features_256)
        if self.patch_filter_params != None:
            contains_tissue_256 = self.filter_tissue(batch_256)
            features_mean256 = features_256[np.where(contains_tissue_256)].mean(dim=0).unsqueeze(dim=0)
        else:
            features_mean256 = features_256.mean(dim=0).unsqueeze(dim=0)

        features_grid256 = features_256.reshape(w_256, h_256, 384).transpose(0,1).transpose(0,2).unsqueeze(dim=0)
        features_grid256 = features_grid256.to(self.device4k, non_blocking=True)
        features_4k = self.model4k.forward(features_grid256).detach().cpu()
        features_mean256_4k = torch.cat([features_mean256, features_4k], dim=1)
        
        asset_dict = {
            'features_256': features_256,
            'features_mean256': features_mean256,
            'contains_tissue_256': contains_tissue_256,
            'features_4k': features_4k,
            'features_mean256_4k': features_mean256_4k
        }
        return asset_dict

    def prepare_img_tensor(self, img: torch.Tensor, patch_size=256):
        make_divisble = lambda l, patch_size: (l - (l % patch_size))
        b, c, w, h = img.shape
        load_size = make_divisble(w, patch_size), make_divisble(h, patch_size)
        w_256, h_256 = w // patch_size, h // patch_size
        return transforms.CenterCrop(load_size)(img), w_256, h_256

    def filter_tissue(self, batch_256):
        to_filter = np.array([filterPatch(img, patch_filter_params=self.patch_filter_params) 
                              for img in tensorbatch2im(batch_256)])
        return 1-to_filter
```

```python
from torch.utils.data.dataset import Dataset

class CSVDataset_DigestPath(Dataset):    
    def __init__(self, dataroot, csv_path, transforms_eval):
        self.dataroot = dataroot
        self.csv = pd.read_csv(csv_path)
        self.transforms = transforms_eval
        
    def __getitem__(self, index):
        img_fname = self.csv['img_fname'][index][:-4]
        label = self.csv['label'][index]
        if label == 1:
            img = Image.open(os.path.join(self.dataroot, 'Positive', self.csv['img_fname'][index]))
            mask = Image.open(os.path.join(self.dataroot, 'Positive_Mask', self.csv['mask_fname'][index]))
            mask = transforms.ToTensor()(mask)
        else:
            img = Image.open(os.path.join(self.dataroot, 'Negative', self.csv['img_fname'][index]))
            mask = torch.Tensor([torch.nan]).unsqueeze(dim=0)
        return self.transforms(img), mask, label, img_fname
    
    def __len__(self):
        return self.csv.shape[0]

train = '/media/hdd2/PatchDatasets/MICCAI2019_DigestPath/Colonoscopy_tissue_segment_dataset/Train'
pos = pd.DataFrame(os.listdir(os.path.join(train, 'Positive')), columns=['img_fname'])
pos['mask_fname'] = pos['img_fname'].str[:-4]+'_mask.jpg'
pos['label'] = 1
for mask_fname in pos['mask_fname']:
    assert mask_fname in os.listdir(os.path.join(train, 'Positive_Mask'))
neg = pd.DataFrame(os.listdir(os.path.join(train, 'Negative')), columns=['img_fname'])
neg['mask_fname'] = np.NaN
neg['label'] = 0
df = pd.concat([neg, pos])
df.to_csv(os.path.join(train, 'train.csv'), index=False)
```
